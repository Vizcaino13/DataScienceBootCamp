```python
# #Concatenate negative reviews and positive reviews together in a new column.

df["concatenated_reviews"] = df["Negative_Review"].astype(str) + df["Positive_Review"]

df["concatenated_reviews"][0]
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stemmer = PorterStemmer()

# Function to tokenize, filter, and stem a review
def tokenize_filter_and_stem(review):
    word_tokens = word_tokenize(review)
    filtered_sentence = [w.lower() for w in word_tokens if w.lower() not in stop_words and w.isalpha()]
    stemmed_words = [stemmer.stem(w) for w in filtered_sentence]
    return ' '.join(stemmed_words)

# Apply stemming to the "concatenated_reviews" column
df["stem_reviews"] = df["concatenated_reviews"].apply(tokenize_filter_and_stem)
```

```python
# #result from stemming
df["stem_reviews"][0]
```

```python
# 1) Lower Case that column
# 2) Remove Stop words
# 3) Remove punctuations
#4) Apply Lemmitization/Stemming (Check accuracy with both)
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
# Function to tokenize and filter a review
def tokenize_filter_and_lem(review):
    word_tokens = word_tokenize(review)
    filtered_sentence = [w.lower() for w in word_tokens if w.lower() not in stop_words and w.isalpha()]
    lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_sentence]
    return ' '.join(lemmatized_words)
df["lem_reviews"] = df["concatenated_reviews"].apply(tokenize_filter_and_lem)
```

```python
#result from Lemmitization
df["lem_reviews"][0]
```

```python
# 5) Apply TFIDF to lemmatized reviews
from sklearn.feature_extraction.text import TfidfVectorizer
v = TfidfVectorizer()
X = v.fit_transform(df['lem_reviews'])
```
