```python


import random

game_board=["{}","{}","{}",
       "{}","{}","{}",
       "{}","{}","{}"]

player="x"
game= True
winner= None

name= input("Enter your name:")


def print_board(game_board):

  print("|",game_board[0] + game_board[1]+ game_board[2],"|")
  print("----------")
  print("|",game_board[3] + game_board[4]+ game_board[5],"|")
  print("----------")
  print("|",game_board[6] + game_board[7]+ game_board[8],"|")



def player_choice(game_board):
  enter_num=int(input("enter a number 1-9:"))

  if enter_num >=1 and enter_num <=9 and game_board[enter_num-1] =="{}":
    game_board[enter_num-1] = player
  else:
      print("The other player has chosen that spot, choose another spot !")




def winhorizontial(game_board):
  global winner
  if game_board[0]==game_board[1]==game_board[2] and game_board[1] !="{}":
    winner=game_board[0]
    return True
  elif game_board[3]==game_board[4]==game_board[5] and game_board[3] !="{}":
    winner=game_board[3]
    return True
  elif game_board[6]==game_board[7]==game_board[8] and game_board[6] !="{}":
    winner=game_board[6]
    return True

def windiagn(game_board):
  global winner
  if game_board[0]== game_board[4]==game_board[8] and game_board[0] !="{}":
    winner=game_board[0]
    return True
  elif game_board[2]== game_board[4]==game_board[6] and game_board[2] != "{}":
    winner=game_board[2]
    return True


def winrow(game_board):
  global winner
  if game_board[0]== game_board[3]==game_board[6] and game_board[0] !="{}":
    winner=game_board[0]
    return True
  elif game_board[1]== game_board[4]==game_board[7] and game_board[1] !="{}":
    winner=game_board[1]
    return True
  elif game_board[2]== game_board[5]==game_board[8] and game_board[2] !="{}":
    winner=game_board[2]
    return True



def tie(game_board):
  global game
  if "{}" not in game_board:
    print_board(game_board)
    print("its a tie")
    game= False


def checkwin():
  global game
  if winhorizontial(game_board) or winrow(game_board) or windiagn(game_board):
    print("this player is the winner!:", winner)
    game= False
    exit()


def switchplayer():
  global player
  if player =="x":
    player ="o"
  else:
    player="x"


def bot(game_board):
  while player=="o":
    position =random.randint(0,8)
    if game_board[position]=="{}":

        game_board[position] ="o"
        switchplayer()




while True:
    game_board = ["{}", "{}", "{}",
             "{}", "{}", "{}",
             "{}", "{}", "{}"]
    player = "x"
    winner = None
    game = True

    while game:
        print_board(game_board)
        player_choice(game_board)
        tie(game_board)
        switchplayer()
        bot(game_board)
        checkwin()
        tie(game_board)

    play_again = input("Do you want to play again? Y or N: ")

    if play_again.upper() != "Y":
        print("Thanks for playing!")
        break
```

```python
#This is creating the board for tic tac toe. This is a list with 9 elements in it indexes 0-8. "-" means that specific point is empty

import random #this is to create the computer in the game

game_board=["{}","{}","{}",
       "{}","{}","{}",
       "{}","{}","{}"]
#all of these are global variables
player="x" #variable set as x since x will go first and the variable, This is also to  remember whose turn it is next
game= True #this is used to control the loop, when it is true the game will continue to run. once it is false the loop will stop and it exits the game
winner= None  #this is used to track when theres a winner in the game

name= input("Enter your name:")

#printing the game board
def print_board(game_board):  #responsible for displaying the current state of the Tic Tac Toe game board on the console, and helps for when you have values in the board

  print("|",game_board[0] + game_board[1]+ game_board[2],"|")
  print("----------")
  print("|",game_board[3] + game_board[4]+ game_board[5],"|")
  print("----------")
  print("|",game_board[6] + game_board[7]+ game_board[8],"|")

#function to print the initial state of the game board to the console



#player input
def player_choice(game_board): #creating a function called player_input to take input for the player
  enter_num=int(input("enter a number 1-9:")) #input function to take a number from the player
  #need to trake the imput and see if its a valid input on the board and not occupied by another player on the board
  if enter_num >=1 and enter_num <=9 and game_board[enter_num-1] =="{}": #make sure the input is valid 1-9, and at the position that the player inputted, making sure no player gone there#note the [inp-1] is subtracting the input and putting it at its correct index
    game_board[enter_num-1] = player  #you set that position to the current player ,  this line of code updates the game board to reflect the player's move by placing their symbol ("x" or "o") in the chosen position.
  else: #if the imput was invalid then it says anither player was in the spot
      print("The other player has chosen that spot, choose another spot !")



#check for all types of wins
def winhorizontial(game_board):
  global winner #if we make changes to the scope of this function then it changed the entire varible in the whole code
  if game_board[0]==game_board[1]==game_board[2] and game_board[1] !="{}": #this is checking that if its all x's in all three elements are if there are no empy spaces
    winner=game_board[0]                                   #then it says this is the winner and it doesn't matter what index its at since its all equal
    return True #this breaks out the game loop if theres a winner
  elif game_board[3]==game_board[4]==game_board[5] and game_board[3] !="{}":
    winner=game_board[3]
    return True
  elif game_board[6]==game_board[7]==game_board[8] and game_board[6] !="{}":
    winner=game_board[6]
    return True

def windiagn(game_board):
  global winner
  if game_board[0]== game_board[4]==game_board[8] and game_board[0] !="{}":
    winner=game_board[0]
    return True
  elif game_board[2]== game_board[4]==game_board[6] and game_board[2] != "{}":
    winner=game_board[2]
    return True

#making functions for each win possibility
def winrow(game_board):
  global winner
  if game_board[0]== game_board[3]==game_board[6] and game_board[0] !="{}":
    winner=game_board[0]
    return True
  elif game_board[1]== game_board[4]==game_board[7] and game_board[1] !="{}":
    winner=game_board[1]
    return True
  elif game_board[2]== game_board[5]==game_board[8] and game_board[2] !="{}":
    winner=game_board[2]
    return True


#check to see if its a tie
def tie(game_board):
  global game
  if "{}" not in game_board: #checks if there are not any blank spaces, and prints the game board
    print_board(game_board)
    print("its a tie")
    game= False #this becomes false to end the game


def checkwin():
  global game
  if winhorizontial(game_board) or winrow(game_board) or windiagn(game_board): #this checks how theres a  winner
    print("this player is the winner!:", winner) #prints out the winner x or o
    game= False #ends the game for who wins
    exit()


#switch player In:  your switchplayer function, there's no need to call the board variable in parentheses because this function is solely
#responsible for toggling the currentplayer variable between "x" and "o." It doesn't directly interact with the board variable.
def switchplayer():
  global player
  if player =="x": #this checks if the current player of that position is x then the value gets reassigned to o
    player ="o"
  else:
    player="x" # Change current player to "x" if it was "o"

#computer
def bot(game_board):
  while player=="o": #the bot will start at o
    position =random.randint(0,8) #this is going with the indexs on the board
    if game_board[position]=="{}": #check to see if that position on the board is taken,
            #if statement checks if the randomly chosen position on the board is empty (indicated by "-"). If the condition is true, it means the computer can make a move in that position.
        game_board[position] ="o"
        switchplayer()




while True:
    game_board = ["{}", "{}", "{}",
             "{}", "{}", "{}",
             "{}", "{}", "{}"]
    player = "x"
    winner = None
    game = True

    while game:
        print_board(game_board)
        player_choice(game_board)
        checkwin()
        tie(game_board)
        switchplayer()
        bot(game_board)
        checkwin()
        tie(game_board)

    play_again = input("Do you want to play again? (Y/N): ")

    if play_again.upper() != "Y":
        print("Thanks for playing!")
        break
```
