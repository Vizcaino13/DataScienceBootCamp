```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.target[[10, 25, 50]]
list(iris.target_names)
```

```python
iris
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# Apply XGBOOST WITH IRIS DATASET FOR CLASSIFICATION

df['target'] = iris.target
df.head()
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
X = df.drop(['target','flower_name'], axis='columns')  #sepal length (cm)	sepal width (cm)	petal length (cm)	petal width (cm)
y = df.target # 0 'setosa', 1'versicolor', 2'virginica'
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from xgboost import XGBClassifier

# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
