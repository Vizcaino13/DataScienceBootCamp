```python
# 1) Clean the dataset look and study the dataset for this part

#What i know is wrong with the data

#the dates has seconds
#invoice numbers are written with a c infront
#formatting in stock code is wrong
#quanity is in the negtaives
#missing CustomerIDs and it has a .0 in it
#Stock code has an A in it
#description has na or "?" in it
#remove duplicates
```

```python
#Error noticed: some invoice numbers are written with a c in front

df["InvoiceNo"]= df["InvoiceNo"].str.lstrip("C")

df
```

```python
#Error noticed: Stock code has letters in it

df["StockCode"]= df["StockCode"].str.strip("ABCDEFGHIJKLMNOPQRSTUVWXYZ")

df
```

```python
#Error noticed: the dates was an object and had seconds

from datetime import datetime

# df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%M %d, %Y')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%y %H:%M')
df['InvoiceDate'] = df['InvoiceDate'].dt.floor('D')
df
```

```python
#quanity is in the negtaives

# df["Quantity"].value_counts()
# #figuring out why they are negative
# QA = [-472, -161, -1206, -272]

# result = df[df["Quantity"].isin(QA)]
# print(result)

# I think it is a typo
df["Quantity"]=df["Quantity"].abs()

df
```

```python
#description has na or "?" in it

dA = ["?"]

result = df[df["Description"].isin(dA)]
print(result)
```

```python
#removing duplicates
df = df.drop_duplicates()
df
```

```python
#2) Create a Dictionary with the 10 countries that have the most 'Quantity' ordered except UK. Create a bar chart for this

import matplotlib.pyplot as plt

country_quan= df.groupby("Country")["Quantity"].sum().reset_index()
country_quan = country_quan[country_quan["Country"] != "United Kingdom"]
top_10_countries = country_quan.sort_values(by="Quantity", ascending=False).head(10)
country_dict = top_10_countries.set_index("Country")["Quantity"].to_dict()
print(country_dict)



# Extract keys (countries) and values (quantities) from the dictionary
countries = list(country_dict.keys())
quantities = list(country_dict.values())

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(countries, quantities)
plt.title('Top 10 Countries by Quantity (Excluding UK)')
plt.xlabel('Country')
plt.ylabel('Total Quantity')

plt.show()


```

```python
#3)  Find Total Sale between two dates entered. line chart

import matplotlib.pyplot as plt
import pandas as pd


start_date = pd.to_datetime("2010-12-01")
end_date = pd.to_datetime("2011-07-21")


filtered_df = df[(df['InvoiceDate'] >= start_date) & (df['InvoiceDate'] <= end_date)]


filtered_df['TotalSale'] = filtered_df['Quantity'] * filtered_df['UnitPrice']
total_sale = filtered_df['TotalSale'].sum()

print("Total Sale between", start_date.date(), "and", end_date.date(), ":", total_sale)

plt.figure(figsize=(10, 6))
plt.plot([start_date, end_date], [0, total_sale], marker='o', linestyle='--')
plt.title('Total Sales Between Two Dates')
plt.xlabel('Dates')
plt.ylabel('Total Sales')
plt.xticks(rotation=70)
plt.show()

```

```python
#4) Find each Countries Total SALE to find which Countries is highest potential customer.
#do a bar chat for this sale on y axis and countries on x axis

df["Total Sales"] = df["Quantity"] * df["UnitPrice"]
country_sales = df.groupby("Country")["Total Sales"].sum().reset_index()
print(country_sales)

print("===================================================================")
max_sales = country_sales["Total Sales"].max()
max_sales_countries = country_sales[country_sales["Total Sales"] == max_sales]
print("Country(s) with the maximum sales:", max_sales_countries)


x = country_sales["Country"]
y = country_sales["Total Sales"]

# Create the bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(x, y, color='blue')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.title('Total Sales by Country')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability # Ensure labels are displayed properly

# Show the plot
plt.show()
```

```python
#5) Which Customer Id has highest purchase in quantity and total and find out the country of it

df["Total"] = df["Quantity"] * df["UnitPrice"]
customer_sales = df.groupby(["CustomerID", "Country"])["Total"].agg(['sum', 'count']).reset_index()
print(customer_sales)

max_cust= customer_sales[customer_sales['sum'] == customer_sales['sum'].max()]

print(max_cust)
```

```python
#6) Which Customer Id has lowest purchase in quantity and total and find out the country of it

df["Total"] = df["Quantity"] * df["UnitPrice"]
customer_sales = df.groupby(["CustomerID", "Country"])["Total"].agg(['sum', 'count']).reset_index()
print(customer_sales)

min_cust= customer_sales[customer_sales['sum'] == customer_sales['sum'].min()]
print(min_cust)
```

```python
#7) Find the biggest order In Sale (Cost) which country was the order from and find total purchase (Cost) done by that customer ID
df["Total"] = df["Quantity"] * df["UnitPrice"]

order_sales = df.groupby(["InvoiceNo", "CustomerID", "Country"])["Total"].agg(['sum', 'count']).reset_index()

max_order= order_sales[order_sales['sum'] == order_sales['sum'].max()]
print(max_order)

customer_id = max_order['CustomerID'].iloc[0]
total_purchase_by_customer = df[df['CustomerID'] == customer_id]['Total'].sum()

print(f"Total purchase (cost) done by Customer ID {customer_id}: {total_purchase_by_customer}")
```

```python
#8) Find how many different customers are available for each country save the count in dictionary also return countries with highest and lowest customer count
#bar chart for customer count. customer count on y axis and countries on x axis
customer_count= df.groupby("Country")["CustomerID"].nunique().reset_index()
customer_count_dict = customer_count.set_index("Country")["CustomerID"].to_dict()
print(customer_count_dict)

max_cus= customer_count[customer_count["CustomerID"] == customer_count["CustomerID"].max()]
print("This is the highest customer count",max_cus)

min_cus= customer_count[customer_count["CustomerID"] == customer_count["CustomerID"].min()]
print("This is the lowest customer count",min_cus)



# Extract keys (countries) and values (quantities) from the dictionary
countries1 = list(customer_count_dict.keys())
cus_count = list(customer_count_dict.values())

# Create a bar chart
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.barh(countries1, cus_count)
plt.title('Customers in each country')
plt.xlabel('Total Customers')
plt.ylabel('Country')
plt.show()

```
