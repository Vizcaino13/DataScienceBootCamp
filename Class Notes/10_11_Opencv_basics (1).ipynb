
#Difference between Blurring vs Erosion
#Blurring
* involves averaging the pixel values in an image
* used to reduce noise in an image
* removes high frequency content (eg: noise, edges) from the image

## Types of blurring
* Averaging
* Gaussian Blurring
* Median Blurring
* Bilateral Filtering


## Eroision
* Erosion is the morphological operation that is performed to reduce the size of the foreground object.
* The boundary of the foreign object is slowly eroded. Erosion has many applications in image editing and transformations, and erosion shrinks the image pixels.
* Pixels on object boundaries are also removed.
