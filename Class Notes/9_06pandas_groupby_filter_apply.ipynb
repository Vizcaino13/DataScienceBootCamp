```python
import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
df_users = pd.read_csv(url, sep='|')
df_users
```

```python
df_users.columns
```

```python
df_users.occupation.unique() #this gives all the unique ocupations
```

```python
df_users.head() #gives all the top records in the data
```

```python
df_users.tail()
```

```python
df_users.groupby("occupation").sum()#when using groupby function there should be one groupby associated with it, it gives the sum of all
```

```python
df_users.groupby("occupation").count()
```

```python
df_users.groupby("occupation").age.count() #this is grouping the data by age and ocupation and counting the ages
```

```python
df2 = df_users.groupby(['occupation','gender']).count().reset_index()
print(df2)
```

```python
df2 = df_users.groupby(by=['gender'], dropna=False).sum()
print(df2)


# This code takes a DataFrame (table) called df_users, groups its data by the 'gender' column, and calculates the sum of the numeric values in each group.
# The dropna=False parameter means it doesn't exclude missing (NaN) values when summing.
# Finally, it prints the result, which shows the sum of values for each gender category.
```

```python
result = df_users.groupby('occupation')['age'].aggregate(['min','max'])
#this is printing the minimum and maximum age of people in certain occupations
print(result)
```

```python
result = df_users.groupby(['occupation', 'gender']).count()

print(result)
```

```python
result = df_users.query("occupation=='homemaker'").groupby(["occupation","gender"]).count()
#this is giving a count of all gender on the occupation of homemaker
print(result)
```

```python
df_homemaker = df_users[df_users['occupation'] == 'homemaker'].groupby('gender').count()
#this is giving a count of all gender on the occupation of homemaker
df_homemaker
```

```python
df_lawyer = df_users[df_users['occupation'] == 'lawyer'].groupby('age').aggregate(["count","min","max"])
df_lawyer
```
