```python
#merging data , when merging there shoukd be common columns inbetween or it should have the same structure

raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])
```

```python
data_new = pd.concat([data1, data2])
data_new
```

```python
data_new_row = pd.concat([data1, data2], axis=0)
data_new_row
#whatever data from data 1 and data2 is there, this is based on row
```

```python
#concatinate on column
# this is based on column
data_new_col=pd.concat([data1,data2],  axis=1)
data_new_col
```

```python
pd.merge(data_new, data3, on="subject_id") #this is the common value inbetween
```
