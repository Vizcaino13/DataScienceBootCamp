```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="/content/supermarket_sales_data.xlsx"
mart= pd.read_excel(path)
mart
```

```python
mart.head(10)
```

```python
# **3.Create a basic COUNTPLOT to get number of transaction for each of the product line and change the figure size**

plt.figure(figsize=(15,5))
sns.countplot(x='Product line', data=mart)
```

```python
# **4.Make it horizontal bar plot**

sns.countplot(y ="Product line", data=mart)
```

```python
# **5.Add hue to get the count on two categories i.e. Product line and Gender**
plt.figure(figsize=(15,5))
```

```python

sns.countplot(x ="Product line", hue="Gender", data=mart)
```

```python
# **6.Using different color palette**

plt.figure(figsize=(15,5))

sns.countplot(x="Product line",hue="Gender",data=mart,palette='PuBuGn_r')
```

```python
# **7.Change Style using facecolor, linewidth and edge color**

plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart
             ,facecolor = (0,0,0,0)
             ,linewidth=5
             ,edgecolor=sns.color_palette('dark',3))
```

```python
#create a basic barplot
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=mart)
```

```python
#**plot the bar in given order**

plt.figure(figsize=(15,5))

sns.barplot(x='Product line',
            y='Total',
            hue='Gender',
            data=mart,
            order=['Electronic accessories', 'Fashion accessories','Food and beverages', 'Health and beauty', 'Home and lifestyle','Sports and travel'],
            hue_order=['Male','Female'])

x=mart['Product line'].sort_values()
x.unique()

plt.show()
```

```python
#**Add CAP on given Error Bar**

plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic accessories', 'Fashion accessories','Food and beverages', 'Health and beauty', 'Home and lifestyle','Sports and travel']
           ,hue_order=['Male','Female']
           ,capsize=0.2)
```
