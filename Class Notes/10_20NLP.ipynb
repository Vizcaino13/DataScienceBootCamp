```python
#2) Property with best review_score overall and least score overall,( Hotel_Name,  Reviewer_Score)
#finding the min and max average of the properties

overall_score= df.groupby(df["Hotel_Name"])["Reviewer_Score"].mean()

print("Property with best review_score overall:", overall_score.idxmax(), overall_score.max())
print("Property with best review_score overall:", overall_score.idxmin(), overall_score.min())
```
