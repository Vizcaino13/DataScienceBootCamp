Object-oriented programming (OOP) is a computer programming model that organizes software design around data, or objects, rather than functions and logic. An object can be defined as a data field that has unique attributes and behavior.

Objectoriented programming: Anything(an object) a compuer program model that organizes the data in like a blue print. so the class is the blueprint and the ojects are the copies of it. object is instance of a class. In Python's OOP, you create these objects as classes. A class is like a blueprint that describes how an object should look and behave. For example, you can have a "Car" class that defines what a car is, including its color, brand, and how it can start or stop.

```python
class universities:
  def __init__(self, name, year_created, location, acceptance_rate):
    self.name=name
    self.year_created=year_created
    self.location=location
    self.acceptance_rate=acceptance_rate

  def Uni_info(self):
     print("this is the school name:", self.name, "this is the year of creation:",self.year_created,
           "this is the location:", self.location,
           "this is the acceptance rate :", self.acceptance_rate,"%")
```

```python
class car:
  def __init__(self,name,model,fuel,cc,cyclinder):
    self.name=name
    self.model=model
    self.fuel=fuel
    self.cc=cc
    self.cyclinder=cyclinder

  def car_details(self):
    print(self.name,self.model,self.fuel,self.cc,self.cyclinder)
```
